Let's search for Pax6 (gene ID: 43812) in the Drosophila melanogaster on NCBI: https://www.ncbi.nlm.nih.gov/gene/?term=43812

Let's get its sequence in GeneBank: https://www.ncbi.nlm.nih.gov/nuccore/NC_004353.4?from=697689&to=721173&report=genbank

And then use the link to blast it: https://blast.ncbi.nlm.nih.gov/Blast.cgi

## Let's do it through an API from BioPython (entrez)

- https://biopython.org/docs/latest/Tutorial/chapter_entrez.html

```python
from Bio import Entrez, SeqIO

Entrez.email = "lenard.szantho@oist.jp"  # since 2010 it's mandatory to give an email address to use NCBI

handle = Entrez.esearch(db="protein", term="eyeless[gene] AND Drosophila melanogaster[orgn]", retmax=5)
record = Entrez.read(handle)
handle.close()

protein_ids = record["IdList"]
print(f"Found protein IDs: {protein_ids}")

# Fetch the sequence - using NP_524628 (eyeless isoform A) directly for reproducibility
handle = Entrez.efetch(db="protein", id="NP_524628.2", rettype="fasta", retmode="text")
drosophila_pax6 = SeqIO.read(handle, "fasta")
handle.close()

print(f"Downloaded: {drosophila_pax6.description}")
print(f"Length: {len(drosophila_pax6.seq)} aa")

SeqIO.write(drosophila_pax6, "drosophila_pax6.fasta", "fasta")
```

### BLAST through API



```python
from Bio.Blast import NCBIWWW, NCBIXML
import time

def blast_organism(query_seq, organism, description, max_hits=1):
    """BLAST query against specific organism and return top hit(s)"""
    print(f"\nBLASTing against {description}...")

    result_handle = NCBIWWW.qblast(
        program="blastp",
        database="nr",
        sequence=str(query_seq),
        entrez_query=f"{organism}[orgn]",
        hitlist_size=max_hits,
        expect=1e-10
    )

    blast_record = NCBIXML.read(result_handle)
    result_handle.close()

    hits = []
    for alignment in blast_record.alignments[:max_hits]:
        hsp = alignment.hsps[0]
        print(f"  Hit: {alignment.title[:80]}...")
        print(f"  E-value: {hsp.expect}, Score: {hsp.score}, Identity: {hsp.identities}/{hsp.align_length}")

        accession = alignment.accession
        hits.append({
            'accession': accession,
            'title': alignment.title,
            'e_value': hsp.expect,
            'identity': hsp.identities / hsp.align_length
        })

    return hits
```

```python
organisms = [
    ("Homo sapiens", "Human"),
    ("Mus musculus", "Mouse"),
    ("Octopus bimaculoides", "Octopus"),  # California two-spot octopus (sequenced genome)
    ("Doryteuthis pealeii", "Squid"),     # Longfin squid (well-studied)
]

all_hits = {}
for org_name, common_name in organisms:
    hits = blast_organism(drosophila_pax6.seq, org_name, common_name)
    all_hits[common_name] = hits
    time.sleep(5)  # Be nice to NCBI

```

```python
print("\nDownloading homolog sequences...")

sequences = [drosophila_pax6]  # Start with our query

for common_name, hits in all_hits.items():
    if hits:
        acc = hits[0]['accession']
        print(f"Fetching {common_name}: {acc}")

        handle = Entrez.efetch(db="protein", id=acc, rettype="fasta", retmode="text")
        seq_record = SeqIO.read(handle, "fasta")
        handle.close()

        seq_record.id = common_name
        seq_record.description = f"{common_name} Pax6 ({acc})"
        sequences.append(seq_record)

        time.sleep(1)

SeqIO.write(sequences, "pax6_homologs.fasta", "fasta")
print(f"\nSaved {len(sequences)} sequences to pax6_homologs.fasta")
```

### Get multiple Pax family members and build a tree

```python
pax_family = {
    # Pax6 orthologs
    "Dmel_eyeless": "NP_524628.2",
    "Dmel_toy": "NP_524642.1",        # twin of eyeless (Pax6 paralog in flies)
    "Human_PAX6": "NP_000271.1",
    "Mouse_Pax6": "NP_038655.1", 
    "Human_PAX2": "NP_000269.1",      # Different Pax subfamily
    "Human_PAX7": "NP_002575.1",      # Different Pax subfamily
    "Octopus_XP014787872": "XP_014787872.1",
    "Octopus_XP014774742": "XP_014774742.1",  # Try alternate
    "Squid_Tomarev": "AAC26141.1",            # Original 1997 paper
    "Squid_ANN46810": "ANN46810.1",           # What BLAST found
}

sequences_family = []
for name, acc in pax_family.items():
    try:
        handle = Entrez.efetch(db="protein", id=acc, rettype="fasta", retmode="text")
        rec = SeqIO.read(handle, "fasta")
        handle.close()
        rec.id = name
        rec.description = name
        sequences_family.append(rec)
        print(f"Success {name}: {len(rec.seq)} aa")
    except Exception as e:
        print(f"Error {name}: {e}")
    time.sleep(1)

SeqIO.write(sequences_family, "pax_family.fasta", "fasta")
```
